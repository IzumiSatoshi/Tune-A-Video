```python
from moviepy.editor import *
import decord
```

```python
clip = VideoFileClip("../data/yor.mp4")
```

```python
txt_clip = TextClip("original", fontsize=40, color='white')
txt_clip = txt_clip.set_pos('center').set_duration(5)
video = CompositeVideoClip([clip, txt_clip])
video.write_videofile("../data/out/tmp.mp4")
```

```python
video.ipython_display()
```
