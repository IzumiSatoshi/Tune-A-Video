```python
posi = "masterpiece, best quality, 8k, detailed anime style of hatsune miku"
nega = "3d, flat shading, flat color, retro style, 1980s, 1990s, 2000s, 2005 2006 2007 2008 2009 2010 2011 2012 2013 2014 2015 2016 2017 2018 2019, lowres, bad anatomy, bad hands, text, error, missing fingers, extra digit, fewer digits, cropped, worst quality, low quality, normal quality, jpeg artifacts, signature, watermark, username, blurry, artist name, inaccurate limb"
strength = 0.5
seed = 0
fps = 8
```

```python
# I2I
pipe = StableDiffusionImg2ImgPipeline.from_pretrained(
    '../checkpoints/waifu-diffusion',
    torch_dtype = torch.float16 
).to("cuda")
pipe.safety_checker = None
```

```python
video = decord.VideoReader("../data/tilting_head/tilting_head.mp4")
```

```python
original_np_list = list(video[:].asnumpy())
result_img_np_list = []
for frame_np in original_np_list:
    init_img = Image.fromarray(frame_np)
    torch.manual_seed(seed)
    result_img = pipe(
        prompt=posi,
        negative_prompt=nega,
        strength=strength,
        image=init_img,
    ).images[0]
    result_img_np = np.asarray(result_img)
    result_img_np_list.append(result_img_np)
```

```python
result_txt = "i2i strength 0.5"

raw_original_clip = editor.ImageSequenceClip(original_np_list, fps=fps)
raw_result_clip = editor.ImageSequenceClip(result_img_np_list, fps=fps)

original_txt_clip = editor.TextClip("original", fontsize=40, color="black", bg_color="bisque")
original_txt_clip = original_txt_clip.set_duration(raw_original_clip.duration)
original_txt_clip = original_txt_clip.set_position(("center", "top"))

result_txt_clip = editor.TextClip(result_txt, fontsize=40, color="black", bg_color="bisque")
result_txt_clip = result_txt_clip.set_duration(raw_result_clip.duration)
result_txt_clip = result_txt_clip.set_position(("center", "top"))

original_clip = editor.CompositeVideoClip([raw_original_clip, original_txt_clip])
result_clip = editor.CompositeVideoClip([raw_result_clip, result_txt_clip])

final_clip = editor.clips_array([[original_clip, result_clip]])
final_clip.ipython_display()
```

```python
raw_result_clip.write_videofile("../data/out/i2i_strength_05.mp4")
```
