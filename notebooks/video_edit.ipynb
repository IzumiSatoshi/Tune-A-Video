```python
import sys
sys.path.append("../")

from moviepy import editor
import numpy as np
from tuneavideo.util import save_videos_grid, read_gif, save_as_gif

import decord
import numpy as np
from PIL import Image
from glob import glob
decord.bridge.set_bridge('torch')
```

```python
def add_label(clip, text):
    text_clip = editor.TextClip(text, fontsize=40, color="black", bg_color="bisque")
    text_clip = text_clip.set_duration(clip.duration)
    text_clip = text_clip.set_position(("center", "top"))

    clip = editor.CompositeVideoClip([clip, text_clip])
    return clip

def gif_to_clip(gif, fps=8):
    gif_np = [np.asarray(image) for image in gif]
    clip = editor.ImageSequenceClip(gif_np, fps=fps)
    return clip
```

```python
fps = 8
```

```python
video1_path = "../data/out/dancing_firefighter/fix_seed_false.mp4"
video2_path = "../data/out/dancing_firefighter/fix_seed_true.mp4"

pose_path = "../data/woman_dance/openpose.mp4"

video1 = editor.VideoFileClip(video1_path)
video2 = editor.VideoFileClip(video2_path)
v = decord.VideoReader(pose_path)
pose = [i.numpy() for i in v[:16]]

pose = editor.ImageSequenceClip(pose, fps=8)

video1 = add_label(video1, "random seed")
video2 = add_label(video2, "fix seed")
pose = add_label(pose, "openpose")

final = editor.clips_array([[pose, video1, video2]])
final.ipython_display()
```
