```python
import sys
sys.path.append("../")

from moviepy import editor
import numpy as np
from tuneavideo.util import save_videos_grid, read_gif, save_as_gif
```

```python
def add_label(clip, text):
    text_clip = editor.TextClip(text, fontsize=40, color="black", bg_color="bisque")
    text_clip = text_clip.set_duration(clip.duration)
    text_clip = text_clip.set_position(("center", "top"))

    clip = editor.CompositeVideoClip([clip, text_clip])
    return clip

def gif_to_clip(gif, fps=8):
    gif_np = [np.asarray(image) for image in gif]
    clip = editor.ImageSequenceClip(gif_np, fps=fps)
    return clip
```

```python
fps = 8
```

```python
video_400 = gif_to_clip(read_gif("../outputs/yurucamp_stamp/1qkv2qkv_9600steps/samples/sample-50.gif"))
video_4000 = gif_to_clip(read_gif("../outputs/yurucamp_stamp/1qkv2qkv_9600steps/samples/sample-500.gif"))
video_10400 = gif_to_clip(read_gif("../outputs/yurucamp_stamp/1qkv2qkv_9600steps/samples/sample-1300.gif"))

video_400 = add_label(video_400, "400 steps")
video_4000 = add_label(video_4000, "4000 steps")
video_10400 = add_label(video_10400, "10400 steps")

final = editor.clips_array([
    [video_400],
    [video_4000],
    [video_10400],
])
final.ipython_display()
```

```python
video1 = editor.VideoFileClip("../data/out/unfrozen_kv_txt_kv.mp4")
video2 = editor.VideoFileClip("../data/out/3e6.mp4")

video1 = add_label(video1, "3e5")
video2 = add_label(video2, "3e6")

final = editor.clips_array([[video1, video2]])
final.ipython_display()
```

```python
v2v_gif = read_gif("../data/tilting_head/out/v2v_strength_07.gif")
i2i_gif = read_gif("../data/tilting_head/out/i2i_strength_03.gif")

v2v_clip = editor.ImageSequenceClip([np.asarray(i) for i in v2v_gif], fps=fps)
i2i_clip = editor.ImageSequenceClip([np.asarray(i) for i in i2i_gif], fps=fps)
original_clip = editor.VideoFileClip("../data/tilting_head/tilting_head.mp4")

v2v_clip_labeled = add_label(v2v_clip, "v2v strength 0.7")
i2i_clip_labeled = add_label(i2i_clip, "i2i strength 0.3")
original_clip_labeled = add_label(original_clip, "original")

final_clip = editor.clips_array([[original_clip_labeled, i2i_clip_labeled, v2v_clip_labeled]])
final_clip.ipython_display()
```
